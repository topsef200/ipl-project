```python
teams=['Rajasthan Royals', 'Royal Challengers Bangalore',
       'Sunrisers Hyderabad', 'Delhi Capitals', 'Chennai Super Kings',
       'Gujarat Titans', 'Lucknow Super Giants', 'Kolkata Knight Riders',
       'Punjab Kings', 'Mumbai Indians']
```

```python
match_df.columns
```

```python
match_df['Team1']=match_df['Team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['Team2']=match_df['Team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['Team1']=match_df['Team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['Team2']=match_df['Team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

match_df['Team1']=match_df['Team1'].str.replace('Kings XI Punjab','Punjab Kings')
match_df['Team2']=match_df['Team2'].str.replace('Kings XI Punjab','Punjab Kings')

match_df['Team1']=match_df['Team1'].str.replace('Gujarat Lions','Gujarat Titans')
match_df['Team2']=match_df['Team2'].str.replace('Gujarat Lions','Gujarat Titans')
```

```python
match_df=match_df[match_df['Team1'].isin(teams)]
match_df=match_df[match_df['Team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df=match_df[['ID','City','WinningTeam','total_run','Team1','Team2']]
```

```python
delivery_df=match_df.merge(delivery,on='ID')
```

```python
delivery_df
```

```python
delivery_df=delivery_df[delivery_df['innings']==2]
```

```python
delivery_df.shape
```

```python
delivery_df.to_csv('c2.csv')
```

```python
delivery_df['current_score']=delivery_df.groupby('ID').cumsum()['total_run_y']
```

```python
delivery_df
```

```python
delivery_df['runs_left']=delivery_df['total_run_x']-delivery_df['current_score']
```

```python
delivery_df.sort_values(by=['ID','overs'])
```

```python
delivery_df['balls_left']=120-(delivery_df['overs']*6+delivery_df['ballnumber'])
```

```python
delivery_df.head()
```

```python
wickets=delivery_df.groupby('ID').cumsum()['isWicketDelivery'].values
delivery_df['wickets']=10-wickets
```

```python
delivery_df
```

```python
#crr=runs/overs
delivery_df['crr']=(delivery_df['current_score']*6)/(120-delivery_df['balls_left'])
```

```python
delivery_df['rrr']=(delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
delivery_df
```

```python
def result(row):
    return 1 if row['BattingTeam']==row['WinningTeam'] else 0
```

```python
delivery_df.apply(result,axis=1)
```

```python
delivery_df['result']=delivery_df.apply(result,axis=1)
```

```python
def BowlingTeam(row):
    return row['Team1'] if row['Team2']==row['BattingTeam'] else row['Team2']
```

```python
delivery_df['BowlingTeam']=delivery_df.apply(BowlingTeam,axis=1)
```

```python
final_df.dropna(inplace=True)
```

```python
final_df=final_df[final_df['balls_left']!=0] #so that rrr will not be infinite
```

```python
X=final_df.iloc[:,:-1]
y=final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
#as 1st 3 columns are in string, we need one hot encoding
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf=ColumnTransformer([('trf',OneHotEncoder(sparse=False,drop='first'),['BattingTeam','BowlingTeam','City'])],remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline(steps=[('step1',trf),('step2',LogisticRegression(solver='liblinear'))])
```

```python
pipe.fit(X_train,y_train)
```

```python
X_train.describe()
```

```python
y_pred=pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(X_test)[7]
```
